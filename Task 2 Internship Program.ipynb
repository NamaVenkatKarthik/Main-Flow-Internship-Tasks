```python
import pandas as pd
```

```python
# importing the csv file
data=pd.read_csv('01.Data Cleaning and Preprocessing.csv')
data
```

```python
# Checking Null values..
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
# Filling null values..
df=data.fillna(value=0)
df
```

```python
df.isnull().sum().sum()
```

```python
# Filling Null values with Previous value...
df2=data.fillna(method='pad')
df2
```

```python
df2.isnull().sum().sum()
```

```python
# Filling Null values with the next value..
df3=data.fillna(method='bfill')
df3
```

```python
df3.isnull().sum().sum()
```
